## Graph the data

```python
from matplotlib.patches import Circle

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(6, 7))



# Draw the penguin body as a large black circle
body = Circle((0, 0), radius=4, facecolor='black', edgecolor='white')
ax.add_patch(body)

# Draw the penguin head as a smaller black circle
head = Circle((0, 3), radius=2, facecolor='black', edgecolor='white')
ax.add_patch(head)

# Draw the penguin eyes as two white circles
left_eye = Circle((-1, 3.5), radius=0.3, facecolor='white')
right_eye = Circle((1, 3.5), radius=0.3, facecolor='white')
ax.add_patch(left_eye)
ax.add_patch(right_eye)

# Draw the pupils as two black circles within the eyes
left_pupil = Circle((-1, 3.5), radius=0.1, facecolor='black')
right_pupil = Circle((1, 3.5), radius=0.1, facecolor='black')
ax.add_patch(left_pupil)
ax.add_patch(right_pupil)

# Draw the penguin beak as an orange triangle
beak = plt.Polygon([(-0.5, 3), (0.5, 3), (0, 2)], facecolor='orange')
ax.add_patch(beak)

# Draw the penguin feet as two orange triangles
left_foot = plt.Polygon([(-1, -4), (-0.5, -3), (-1.5, -3)], facecolor='orange')
right_foot = plt.Polygon([(1, -4), (0.5, -3), (1.5, -3)], facecolor='orange')
ax.add_patch(left_foot)
ax.add_patch(right_foot)

# Set the axis limits and title
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 6)
ax.set_title('Simple Penguin Drawing')

# Display the plot
plt.show()

# This one was totally made with chatGPT
```

```python

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(21, 5))

colors = [
    '#FF5733', # vivid orange
    '#C70039', # crimson red
    '#900C3F', # dark red
    '#FFC300', # lemon yellow
    '#9B59B6', # deep purple
    '#3498DB', # bright blue
    '#27AE60', # forest green
    '#F39C12', # bright orange
    '#1ABC9C', # teal green
    '#E67E22', # pumpkin orange
]

for i, color in enumerate(colors):
    x = np.linspace(0, 10, 10*(1+i))
    y = np.sin(x)
    plt.plot(x+(12*i), y, color=color)
    
ax.set_title('Simple Color Play On Graph')
ax.set_facecolor('#F3')

plt.show()
```
