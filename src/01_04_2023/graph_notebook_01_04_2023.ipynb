```python
# Standard imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
import ssl
import requests

%matplotlib inline
```

## Graph the data

```python
# Define the four points as a list of tuples
points = [(2, 8), (5, 15), (6, 17), (8, 23)]

# Create a new figure and axis
fig, ax = plt.subplots()

# Create the scatter plot of the points
ax.scatter([p[0] for p in points], [p[1] for p in points], color='blue')


# Add an arrow and label for each point
for i, p in enumerate(points):
    dx = 1.5  # Arrow length in x direction
    dy = 0.8  # Arrow length in y direction
    ax.arrow(p[0]+0.2, p[1]+0.2, dx, dy, head_width=0.2, head_length=0.5, fc='red', ec='black')
    ax.annotate(f'Point {i+1}', (p[0]+dx+0.8, p[1]+dy+0.2), color='green')

# Set the axis limits and title
ax.set_xlim(0, 12)
ax.set_ylim(0, 27)
ax.set_title('Scatter Plot with Arrows and Labels')

# Display the plot
plt.show()
```
