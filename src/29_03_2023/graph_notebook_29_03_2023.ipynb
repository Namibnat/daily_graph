# Project: Graph of a function

```python
# Standard imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
import ssl
import requests

%matplotlib inline
```

## Graph the data

```python
# Define the function to plot
def f(x):
    return np.sin(x)

# Specify the range of x-values to plot
# Use NumPy to generate an array of 100 equally spaced values between -π and π, inclusive
x_values = np.linspace(-(2*np.pi), (2*np.pi), 200)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x_values, f(x_values), color='#388338')

# Set the axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Graph of y = sin(x)')

# Add the x-axis and y-axis lines. axhline -> 'add x horizontal line'
ax.axhline(y=0, color='black', linewidth=0.5)
ax.axvline(x=0, color='black', linewidth=0.5)


# Remove the top and right spines
for spin in ('top', 'right', 'bottom', 'left'):
    ax.spines[spin].set_visible(False)

# Display the plot
plt.show()
```
