```python
# Standard imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
%matplotlib inline
```

## Get Data

Data source: https://www.kaggle.com/datasets/muthuj7/weather-dataset

```python
path = os.path.join('/', *os.getcwd().split('/')[1:-2], '.data')

data = pd.read_csv(os.path.join(path, 'weather.csv'))
data.shape
```

## Clean and organize dataframe

```python
data.columns
```

```python
data_at_9 = data[['WindSpeed9am', 'Temp9am']]
data_at_9.head()
```

## Graph the data

```python

# create the graph
plt.plot(data_at_9['Temp9am'], label='Temperature at 9:00')
plt.plot(data_at_9['WindSpeed9am'], label='Wind Speed at 9:00')
plt.title("Temperature and windspeeds at 9:00")
plt.legend()

plt.show()
```

## Conclusions


The data suggests...There's some vague relationship between temperatures and wind speed.  Quick graph on a busy day.
