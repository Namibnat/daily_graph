```python
# Standard imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
import ssl
import requests

%matplotlib inline


# Extra for today
import random
```

## Get Data

```python
# Just fake data today
# Seed the random number generator for reproducibility
np.random.seed(42)

# Generate 50 random x and y values
x_values = np.random.randint(0, 100, 50)
x_values_sorted = np.sort(x_values)

y_values = np.random.randint(0, 20, 50)

# Create a pandas DataFrame with the random x and y values
data = pd.DataFrame({'x': x_values_sorted, 'y': y_values})

# Display the DataFrame
data.head()
```

## Graph the data

```python
# Imports specific to this graph
import matplotlib.image as mpimg
```

```python
# Graph image for background
image = mpimg.imread('/Users/sphen/Desktop/penguin_bg.jpeg')
```

```python
fig, ax = plt.subplots()

ax.imshow(
    image,
    aspect='auto',
    extent=[data['x'].min(),
            data['x'].max(),
            data['y'].min(),
            data['y'].max()],
    alpha=0.5, zorder=-1)


# Place data on the map
ax.plot(data['x'], data['y'], color='#FF7F50')


plt.show()
```
