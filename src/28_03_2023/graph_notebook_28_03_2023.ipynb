```python
# Standard imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
import ssl
import requests

%matplotlib inline
```

## Graph the data

Data:

- Russia - 17,098,242 square kilometers
- Canada - 9,984,670 square kilometers
- China - 9,596,961 square kilometers
- United States - 9,525,067 square kilometers

Reference on which the code is based:
https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html


```python
fig, ax = plt.subplots()

fruits = ['Russia', 'Canada', 'China', 'United States']
counts = [17098242, 9984670, 9596961, 9525067]

bar_colors = ['darkred', 'navy', 'goldenrod', 'sienna']

ax.bar(fruits, counts, color=bar_colors)

# Horizontal line for the average land area of these countries
world_average = np.mean(counts)
ax.axhline(world_average, color='gray', linestyle='--', label='World average')

# Add an annotation for the world average line
ax.annotate('Average Of These Countries', xy=(0.4, world_average), xytext=(50, 10),
            xycoords=('axes fraction', 'data'), textcoords='offset points',
            arrowprops=dict(facecolor='black', arrowstyle='->'))


ax.set_ylabel('Country size (sq km)')
ax.set_title("The size of the world's largest countries")
ax.set_facecolor('linen')

plt.show()
```
