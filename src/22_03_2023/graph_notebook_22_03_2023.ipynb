```python
# Made up land use (This data was just made up by chatGPT since I was short of time)

land_use_distribution = {
    "Agriculture": 42.8,
    "Forestry": 29.6,
    "Urban Development": 19.3,
    "Protected Natural Areas": 8.3
}

print("Name" + ' '*20 + '|' + ' Value')
print('-'*24 + '|' + '-'*6)
for name, value in land_use_distribution.items():
    print(f"{name:<24}|{value:>6}")
```

## Graph the data

```python
# Extract the labels and values from the dictionary
labels = list(land_use_distribution.keys())
sizes = list(land_use_distribution.values())

# Create the pie chart (note to myself, autopct is 'auto percentage')
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Set the title
plt.title("Land Use Distribution")

# Display the chart
plt.show()
```
